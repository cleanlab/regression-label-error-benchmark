```python
import utils
import numpy as np
```

### Generate simulated data

```python
Data = utils.Get_data(1000,.1,3)
training_data = Data.drop(['true_label', 'true_error','given_label'], axis=1)  
label=Data['given_label']
```

### Train model

```python
predictorArgs = {
                    "path"             : "./MyAutogluonModels", 
                    "label"            : 'given_label',
                    "problem_type"     : 'regression'
                }
My_AG_estimator =utils.MyAutoGluonPredictor(**predictorArgs)
My_AG_estimator.fit(training_data,label)
```

### Get predicted y, sigma, and u

```python
Y_pred=My_AG_estimator.predict(model="LightGBM_BAG_L1")
Sigma_pred=My_AG_estimator.predict_sigma(model="LightGBM_BAG_L1")
U_predict=My_AG_estimator.predict_u(training_data, model="LightGBM_BAG_L1")
```

### Get proposed scores

```python
scores_res=abs(Y_pred -Data['given_label']) ## residual as the baseline method
scores_ari=abs(Y_pred -Data['given_label'])/(Sigma_pred+U_predict) ## arithmetic mean 
scores_geo=abs(Y_pred -Data['given_label'])/np.sqrt(Sigma_pred*U_predict) ## geometric mean 
```
